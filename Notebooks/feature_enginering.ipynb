```python
# Unnamed ustunlarni drop qilish
df = df.drop(columns=['Unnamed: 0'])

# Natijani tekshirish
print(df.info())
```

# Fayl ko‘rilishiga nisbatan qancha yuklanganini bildiradi.

```python
df['download_view_ratio'] = df['downloadcount'] / (df['viewcount'] + 1)
```

# Faylning jalb qilish darajasi — ko‘rilish va yuklanishni birlashtiradi.

```python
df['engagement_score'] = df['downloadcount']*0.6 + df['viewcount']*0.4
```

# Bu ustun har bir yozuvga tegishli bo‘lgan teglar (tags) ro‘yxatini o‘z ichiga oladi.

```python
df['num_tags'] = df['tags'].apply(lambda x: len(x.split(',')))
```

# title (sarlavha) uzunligini hisoblaydi.
# subtitle (kichik sarlavha) uzunligini hisoblaydi.

```python
df['title_length'] = df['title'].apply(len)
df['subtitle_length'] = df['subtitle'].apply(len)
```

# creatorname (yaratuvchi) qanchalik mashhur ekanligini ko‘rsatadi.
# ownername (egasi) qanchalik mashhur ekanligini ko‘rsatadi.

```python
creator_counts = df['creatorname'].value_counts().to_dict()
df['creator_popularity'] = df['creatorname'].map(creator_counts)

owner_counts = df['ownername'].value_counts().to_dict()
df['owner_popularity'] = df['ownername'].map(owner_counts)
```

# Quyidagi kod `lastupdated` ustuni orqali vaqtga oid yangi xususiyatlar yaratadi:

```python
df['lastupdated'] = pd.to_datetime(df['lastupdated'])
df['lastupdated_year'] = df['lastupdated'].dt.year
df['lastupdated_month'] = df['lastupdated'].dt.month
df['lastupdated_day'] = df['lastupdated'].dt.day
df['lastupdated_weekday'] = df['lastupdated'].dt.weekday
```
