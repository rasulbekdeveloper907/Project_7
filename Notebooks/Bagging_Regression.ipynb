```python

#df
x=df.drop('Comments',axis=1)
y=df['Comments']
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python

base_learners = [
    ('Linear Regression', LinearRegression()),
    ('Decision Tree', DecisionTreeRegressor(max_depth=5)),
    ('SVR', SVR())
]
```

# Predictions

```python

base_train_preds = []
base_test_preds = []
print("=== Base Models ===")
for name, model in base_learners:
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    base_train_preds.append(y_train_pred)
    base_test_preds.append(y_test_pred)
#Individual
    r2_train = model.score(X_train, y_train)
    r2_test = model.score(X_test, y_test)
    print(f"{name} - R² train: {r2_train:.3f}, test: {r2_test:.3f}")
```

```python
import numpy as np
X_meta_train = np.column_stack(base_train_preds[0])
X_meta_test = np.column_stack(base_test_preds[0])
print(X_meta_train)
print(X_meta_test)
```

# Level -1 (Meta model using predictions)

```python
meta_model = Ridge()
meta_model.fit(X_meta_train, y_train)

y_final_pred = meta_model.predict(X_meta_test)
r2_final = meta_model.score(X_meta_test, y_test)

print("\n=== Meta Model (Stacking) ===")
print(f"Final Predictions: {y_final_pred[:5]}")
print(f"R² test (final output): {r2_final:.3f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# ============================================
# 1) Train-test split
# ============================================
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# ============================================
# 2) Base (level-0) modellari
# ============================================
base_models = [
    ("lr", LinearRegression()),
    ("tree", DecisionTreeRegressor(max_depth=6, random_state=42)),
    ("rf", RandomForestRegressor(n_estimators=200, random_state=42)),
    ("gboost", GradientBoostingRegressor(random_state=42)),
]

# ============================================
# 3) Base modellarning prediktsiyalari → Meta train input
# ============================================
meta_train = []

for name, model in base_models:
    model.fit(X_train, y_train)
    pred = model.predict(X_train)
    meta_train.append(pred)

# model soni = ustunlar, sample soni = qatorlar
X_meta_train = np.column_stack(meta_train)

# ============================================
# 4) Base modellarning test prediktsiyalari → Meta test input
# ============================================
meta_test = []

for name, model in base_models:
    pred = model.predict(X_test)
    meta_test.append(pred)

X_meta_test = np.column_stack(meta_test)

# ============================================
# 5) Meta model (Level-1 model)
# ============================================
meta_model = Ridge()
meta_model.fit(X_meta_train, y_train)

# ============================================
# 6) Final stacking prediktsiya
# ============================================
y_final_pred = meta_model.predict(X_meta_test)

# ============================================
# 7) Baholash
# ============================================
r2_final = r2_score(y_test, y_final_pred)

print("=== META MODEL (STACKING) ===")
print("X_meta_train shape:", X_meta_train.shape)
print("X_meta_test shape:", X_meta_test.shape)
print("Final Predictions:", y_final_pred[:5])
print(f"R² test (final output): {r2_final:.4f}")
```

# Bagging + Classification 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, ExtraTreesClassifier
```

```python
x=df.drop('CategoryName',axis=1)
y=df['CategoryName'].astype(int)
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

# Kop ishlatilanadigan xolatlar

```python
# 1. Random Forest
# ----------------------------
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
# 2. Bagged Decision Trees
# ----------------------------
bag_dt = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    random_state=42
)
bag_dt.fit(X_train, y_train)
y_pred_bagdt = bag_dt.predict(X_test)
print("Bagged Decision Trees Accuracy:", accuracy_score(y_test, y_pred_bagdt))
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
print("Bagged Decision Trees Accuracy:", accuracy_score(y_test, y_pred_dt))
```

```python
# 3. Extra Trees
# ----------------------------
extra_trees = ExtraTreesClassifier(n_estimators=100, random_state=42)
extra_trees.fit(X_train, y_train)
y_pred_extra = extra_trees.predict(X_test)
print("Extra Trees Accuracy:", accuracy_score(y_test, y_pred_extra))
```

# Kam ishlatilanadigan xolatlar

```python
# 4. Bagging KNN
# ----------------------------
bag_knn = BaggingClassifier(
    estimator=KNeighborsClassifier(n_neighbors=3),
    n_estimators=50,
    random_state=42
)
bag_knn.fit(X_train, y_train)
y_pred_knn = bag_knn.predict(X_test)
print("Bagged KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print("Bagged Decision Trees Accuracy:", accuracy_score(y_test, y_pred_knn))
```

```python
# 5. Bagging SVM
# ----------------------------
bag_svm = BaggingClassifier(
    estimator=SVC(kernel='linear', probability=True),
    n_estimators=50,
    random_state=42
)
bag_svm.fit(X_train, y_train)
y_pred_svm = bag_svm.predict(X_test)
print("Bagged SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
```

```python
# 6. Bagging Neural Network
# ----------------------------
bag_nn = BaggingClassifier(
    base_estimator=MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, random_state=42),
    n_estimators=20,
    random_state=42
)
bag_nn.fit(X_train, y_train)
y_pred_nn = bag_nn.predict(X_test)
print("Bagged Neural Network Accuracy:", accuracy_score(y_test, y_pred_nn))
```
