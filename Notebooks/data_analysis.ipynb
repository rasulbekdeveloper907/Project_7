# Eng ko‚Äòp yuklab olingan 10 ta dataset

```python
import plotly.express as px

# 1. Eng ko'p yuklab olingan 10 ta datasetni tanlash
top10_downloads = df.nlargest(10, 'downloadcount')

# 2. Plotly Express bilan bar chart chizish
fig = px.bar(
    top10_downloads,
    x='subtitle',              # X o'qda dataset nomi
    y='downloadcount',         # Y o'qda yuklab olishlar soni
    hover_data=['creatorname', 'url'],  # Hoverda ko'rsatish uchun
    color='downloadcount',     # Ranglar yuklab olishlar soniga mos
    title='üìä Top 10 Most Downloaded Kaggle Datasets',
    labels={'subtitle': 'Dataset', 'downloadcount': 'Download Count'}
)

# 3. X o'qini vertikal qilish (agar uzun nomlar bo'lsa)
fig.update_layout(xaxis_tickangle=-45)

# 4. Grafni ko'rsatish
fig.show()
```

# Top 10 creatorlar bo'yicha dataset soni

```python
import plotly.express as px


top_creators = df['creatorname'].value_counts().nlargest(10).reset_index()
top_creators.columns = ['creatorname', 'dataset_count']

fig_creators = px.bar(
    top_creators,
    x='creatorname',
    y='dataset_count',
    color='dataset_count',
    title='üë§ Top 10 Most Active Kaggle Creators',
    labels={'creatorname': 'Creator', 'dataset_count': 'Number of Datasets'}
)
fig_creators.update_layout(xaxis_tickangle=-45)
fig_creators.show()
```

# Time series analysis ‚Äì Dataset yaratish tendensiyasi

```python
import plotly.express as px
import pandas as pd

# lastupdated ni datetime formatiga o'tkazish
df['lastupdated'] = pd.to_datetime(df['lastupdated'], errors='coerce')

# Datasetlar soni bo'yicha oylar bo'yicha guruhlash
df_time = df.groupby(df['lastupdated'].dt.to_period('M')).size().reset_index(name='dataset_count')
df_time['lastupdated'] = df_time['lastupdated'].dt.to_timestamp()

fig_time = px.line(
    df_time,
    x='lastupdated',
    y='dataset_count',
    title='üìà Number of Kaggle Datasets Over Time',
    labels={'lastupdated': 'Month', 'dataset_count': 'Number of Datasets'}
)
fig_time.show()
```
