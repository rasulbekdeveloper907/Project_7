```python
# Olib tashlanadigan ustunlar ro'yxati
columns_to_drop = [
    'isprivate', 'isfeatured', 'licensename', 'kernelcount', 
    'topiccount', 'votecount', 'currentversionnumber', 'usabilityrating', 
    'files', 'versions', 'isfrozen'
]

# Ustunlarni olib tashlash
df_unique = df_unique.drop(columns=columns_to_drop)
```

```python
df_unique.info()
```

```python
# Tekshiriladigan ustunlar ro'yxati
columns_to_check = [
    'id', 'ref', 'subtitle', 'creatorname', 'creatorurl', 
    'totalbytes', 'url', 'lastupdated', 'downloadcount', 
    'ownername', 'ownerref', 'title', 'viewcount', 'tags'
]

# Har bir ustun bo'yicha unikal qiymatlar soni
unique_counts = df_unique[columns_to_check].nunique()

# Natijani chiqarish
print(unique_counts)
```

```python
# 1. Har bir ustun bo'yicha unikal qiymatlar sonini hisoblash
unique_counts = df_unique.nunique()

# 2. Eng kam unique qiymatga ega ustunni topish
min_unique_col = unique_counts.idxmin()
min_unique_count = unique_counts.min()
print(f"Eng kam unique qiymatga ega ustun: {min_unique_col} ({min_unique_count} unikal qiymat)")

# 3. Shu ustun bo'yicha duplicate qatorlarni olib tashlash
df_filtered = df_unique.drop_duplicates(subset=[min_unique_col]).reset_index(drop=True)

# Natijani tekshirish
print(df_filtered.shape)
print(df_filtered[min_unique_col].nunique())  # endi eng kam unique qiymatga ega ustun bo'yicha son
```
